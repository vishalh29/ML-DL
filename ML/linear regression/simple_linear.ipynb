```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("/workspaces/My-Learning/ML DL NLP/3-Complete Linear Regression/Practicals/height-weight.csv")
df[:5]
```

```python
X = df[["Weight"]]
y = df["Height"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, random_state=42, test_size=0.25)
```

```python
X_train
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
X_test
```

```python
from sklearn.linear_model import LinearRegression
```

```python
regression = LinearRegression()
```

```python
regression.fit(X_train, y_train)
```

```python
print("coefficient", regression.coef_)
print("intercept", regression.intercept_)
```

```python
plt.scatter(X_train, y_train, color='blue')
plt.plot(X_train, regression.predict(X_train), color='red')
```

```python
y_pred = regression.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score, adjusted_rand_score
```

```python
MSE = mean_squared_error(y_test, y_pred)
MAE = mean_absolute_error(y_test, y_pred)
R2 = r2_score(y_test, y_pred)
rmse = np.sqrt(MSE)
adjusted_r2 = 1 - (1-R2)*(len(y_test)-1)/(len(y_test)-X_test.shape[1]-1)

print("Mean Absolute Error:", MAE)
print("Mean Squared Error:", MSE)
print("Root Mean Squared Error:", rmse)
print("R^2 Score:", R2)
print("Adjusted R^2 Score:", adjusted_r2)
```
